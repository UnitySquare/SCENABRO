```python
from transformers import pipeline

summarizer = pipeline("summarization", model="Falconsai/text_summarization")

ARTICLE = """
Returning student Hong Man-seop (Ahn Jae-hong) returned from the military and visited the foot volleyball court in his dream after returning to the university, but the foot volleyball court was removed and disappeared. To solve this problem, he visited the bachelor's school and consulted, but to no avail, and in a conversation with the president hosted by the school, he proudly suggested, "Please make a foot volleyball court," but of course, it was not adopted. While recruiting people for a signature campaign to create a foot volleyball court as the next best option, he naturally opens the door to conversation by meeting with Anna (Hwang Seung-eon), the queen of the campus, but is blocked by Kang Min, a former national soccer player who keeps him in check, and the reason to concede Anna to each other and played a one-on-one foot volleyball match with his car.
The game video at this time becomes UCC and spreads on the school, and foot volleyball fever blows on the school. As a result, as the foot volleyball tournament is held in the school, the story gradually begins to struggle with the Department of Food and Nutrition to win the championship.
"""
print(summarizer(ARTICLE, max_length=230, min_length=30, do_sample=False))
```
