```python
from diffusers import AutoPipelineForText2Image
import torch

my_path = "/content/drive/MyDrive/my_ws/project/Final/sdxl-turbo"  # 직접 다운로드 받은 파일 구글 드라이브

# my_path = "stabilityai/sdxl-turbo"  # 온라인 상에서 다운받아서 쓰기

pipe = AutoPipelineForText2Image.from_pretrained(my_path, torch_dtype=torch.float16, variant="fp16")
pipe.to("cuda")
```

```python
prompt = "cartoon: Man-seop, who returned to school after serving in the military, finds out that the foot volleyball court has disappeared from the school."

image = pipe(prompt=prompt, num_inference_steps=1, guidance_scale=0.0).images[0]

image
```
