```python
from diffusers import DiffusionPipeline, LCMScheduler

pipe = DiffusionPipeline.from_pretrained("Lykon/dreamshaper-7",safety_checker = None,
    requires_safety_checker = False).to("cuda")
pipe.scheduler = LCMScheduler.from_config(pipe.scheduler.config)
pipe.load_lora_weights("latent-consistency/lcm-lora-sdv1-5") #yes, it's a normal LoRA
```

```python
import os
import re
import pandas as pd
from PIL import Image
from tqdm import tqdm
from googletrans import Translator  # 번역

trans = Translator()

# Data Frame Create
df = pd.read_csv('/content/drive/MyDrive/serieson.csv', encoding='cp949', header=None)
df
```

```python
# 이미지 저장 경로 설정
save_path = '/content/drive/MyDrive/images'

# csv 파일 저장을 위한 데이터프레임 생성
df_save = pd.DataFrame(columns=['Image Name', 'Text'])

# 해당 경로에 폴더가 없다면 생성
if not os.path.exists(save_path):
    os.makedirs(save_path)

for i in tqdm(range(len(df))):
    row = re.sub('\.{2,}', '.', df.iloc[i][0])
    row = re.sub('[^가-힣ㄱ-ㅎㅏ-ㅣ, .0-9!?…]', '', row)
    print(row)
    s_list = re.split('(?<=[.?!…])\s', row)
    if len(s_list) > 1:
        for j in range(len(s_list)):
            print(s_list[j])
            results = pipe(
                prompt=trans.translate(s_list[j] ,dest='en', src='ko').text,
                num_inference_steps=5,
                guidance_scale=0.0,
            )
            # 이미지 객체로 변환
            image = results.images[0]

            # 이미지 이름 설정 및 저장
            image_name = f'{i+1}-{j+1}.png'  # 예: 1-1, 1-2, 2-1, 2-2 ...
            image.save(os.path.join(save_path, image_name))

            # 데이터프레임에 데이터 추가
            df_save = df_save.append({'Image Name': image_name, 'Text': s_list[j]}, ignore_index=True)

            # 데이터프레임을 csv 파일로 실시간 저장
            df_save.to_csv('/content/drive/MyDrive/save.csv', index=False, encoding='utf-8-sig')
```
